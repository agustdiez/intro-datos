## Programación en Python - Básicos

### Tipo de datos

Tenemos distintos tipos de datos. Conocer esto resulta importante ya que es muy común querer realizar operaciones con datos no permitidos, por lo que debemos poder entender a qué se deben esos errores y transformar el dato de ser necesario:

- **Números**: `float` (reales) o `integer` (enteros)
- **Text**o: `string`
- **Booleanos**: `True` o `False`. Con este tipo de variables se pueden usar operadores lógicos tales como `and`, `or`, `not`, `xor`. Los booleanos se devuelven al comparar distintos tipos de datos
- **Listas**: colección ordenada y modificables de elementos. Pueden contener, números, strings, booleanos, listas o cualquier otro tipo de dato. **Pueden modificarse**
- **Tuplas**: colección ordenada e inmutable de elementos.**No pueden modificarse**
- **Diccionarios**: se designan con {}. Albergan "keys-values". Son desordenadas y se accede con la clave a los valores albergados.

Con los números se pueden realizar los siguientes tipos de operaciones:

| Operación | Resultado       |
| --------- | --------------- |
|    +      | Suma            |
|    -      | Resta           |
|    *      | Multiplicación  |
|    /      | División        |
|    //     | División entera |
|    \*\*   | Potencia        |
|     %     | Resto           |

Métodos útiles con strings:



1.   `find()`: buscar donde aparece en el string determinado valor (primera ocurrencia)
2.   `upper()`: llevar a mayus
3.   `count()`: cuenta cuántas veces aparece en la cadena
4. `replace()`: reemplazar cadena por otra





```python
#Busco índice donde aparece el 1
str_1 = '00000000000000100000000000000000000'
print(str_1.find('1'))
print('El 0 aparece '+ str(str_1.count('0'))+' veces')

str_2 = 'edificio de propiedad horizontal'
print(str_2.upper())
print(str_2.replace('edificio','casa'))
```

#### Listas

```python
li = [-5, 7, 4, 9, 1, 12, 2]
```

La operación más importante que se hace con las listas es la extracción de sus elementos en distintas posiciones. Tener en cuenta que las listas en Python siempre empiezan con el índice `0`

```python
#Accedo a segundo digito
segundo_elemento = li[1]
print(segundo_elemento)

#Cambio el valor del segundo digito
li[1] = 'Hola'
print(li)
```

Si quiero más de un elemento, se hacen slices usando el `:`
- Si se quieren ciertos elementos, se indica inicio y fin.
- Si se quieren todos desde cierto indice hasta el final se usa [indice:]
- Si se quieren todos desde cierto indice hasta el comienzo se usa [:indice]
- Los indices pueden ser negativos, leyendo de derecha a izquierda.


```python
print(li[1:])
print(li[:4])

#Tambien se puede dar un step
print(li[1:5:2])
```

```python
#El print no almacena variables
b = li[1:4]
```

```python
b
```

Las funciones y métodos a utilizar con listas son las siguientes:

Funciones:
1. `len()`: devuelve número de elementos
2. `sorted()`: devuelve lista ordenada
3. `sum()`: suma de todos los valores
4. `min()`: elemento más chico
5. `max()`: elemento más grande
6. `any()`: devuelve True si tiene un True
7. `all()`: devuelve True si tiene todos True

Métodos:
1. `append()`: agrega elemento al fin de la lista
2. `insert()`: ingresa elemento en cierto índice
3. `remove()`: remueve primer valor en la lista
4. `count()`: devuelve la cantidad de veces que aparece cierto elemento


```python
li_2 = [3 , 7 , 9 , -1 , -5]
print('Lista ordenada: '+str(sorted(li_2)))
print('Suma total: '+str(sum(li_2)))
print('Valor máximo: '+str(max(li_2)))
print('Valor mínimo: '+str(min(li_2)))
print('La lista tiene '+str(len(li_2))+' elementos')
```

```python
#Agrego números impares de 1 a 100 a una lista

impares = []
for i in range(0,100):
  if i%2 == 0:
    pass
  else:
    impares.append(i)

print(impares)
```

#### Diccionarios

A diferencia de estar referenciados por índices como en listas, tuplas o strings, en un diccionario se indexa por ***claves (Keys)***. El acceso se hace a través de la key, que puede ser un número, cadenas, tuplas o cualquier dato.

Los valores de una key son mutables, por lo que pueden reasignarse, como en una lista

```python
alumnos = {'Fernando':[27,'Masculino',10234],'Sebastián':[23,'Masculino',99921],'Catalina':[19,'Femenino',101896]}
```

```python
alumnos
```

```python
alumnos['Catalina']
```

```python
# Accedo a un alumno
alumnos['Catalina']

#Corrijo la edad, mal ingresada
alumnos['Catalina'] = [21,'Femenino',101896]
```

```python
alumnos
```

```python
# Chequeo si una key pertenece al diccionario

'Alberto' in alumnos
```

```python
#Para extraer las keys y values

nombres_alumnos = alumnos.keys()
datos_alumnos = alumnos.values()
```

```python
datos_alumnos
```

## Programación en Python - Librerías

Las librerías son conjuntos de funciones y clases ya definidas que pueden importarse para hacer uso de las mismas de manera sencilla. De alguna forma, al importarlas estamos trayendo código a nuestro script para ser ejecutado aunque no lo veamos directamente

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

Se sugiere importarlas con las abreviaturas populares, de forma de buscar código escrito igual que el nuestro

Si la librería es extensa y se requiere solamente importar un comando o método especifico, se usa el FROM

```python
from math import sin
```

```python
sin(60)
```

Si la importamos con un nombre corto

```python
import math as mt
```

```python
mt.sin(60)
```

NO se recomienda importar todos los comandos de una librería de la forma mostrada debajo, ya que entre librerías pueden existir iguales nombres para funciones o métodos y no quedarían diferenciadas

```python
from math import *
```
