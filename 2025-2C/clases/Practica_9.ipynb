# Aprendizaje no supervisado

El aprendizaje no supervisado es un enfoque de aprendizaje que se aplica cuando se dispone únicamente de un conjunto de variables explicativas $(X_1, X_2, ..., X_p)$ medidas sobre $n$ observaciones, sin una variable de respuesta asociada $(Y)$. A diferencia del aprendizaje supervisado, su objetivo no es la predicción, sino explorar y descubrir patrones ocultos o estructuras interesantes en los datos.



# Clustering
En particular, el clustering es una técnica de aprendizaje no supervisado que agrupa datos en subconjuntos (clusters) de tal manera que los datos dentro de un mismo cluster son más similares entre sí que a los de otros clusters.

Se busca:
* **Cohesión** dentro de un mismo clúster: instancias dentro de un mismo clúster son similares entre sí.
* **Separación** entre distintos clusters: instancias de distintos clusters son diferentes entre sí.

## Algunas aplicaciones de algoritmos de clustering
1. Segmentación de mercado
2. Análisis de datos
3. Reducción de dimensionalidad
4. Ingeniería de atributos
5. Detección de anomalías
6. Aprendizaje semisupervisado
7. Motores de búsqueda
8. Segmentación de imágenes

## Distancias

En clustering, se utilizan diversas medidas de distancia para determinar la similitud o diferencia entre los puntos de datos. Algunas medidas comunes son:

1. **Distancia Euclidiana**:
<br>
\begin{align}
d(x, y) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}
\end{align}
* Es la distancia entre dos puntos en el espacio euclidiano. Es adecuada para datos en los que todas las dimensiones tienen la misma escala.

2. **Distancia de Manhattan**:
<br>
\begin{align}
d(x, y) = \sum_{i=1}^{n} |x_i - y_i|
\end{align}

* Suma de las diferencias absolutas de sus coordenadas.

3. **Distancia de Minkowski**:
<br>
\begin{align}
d(x, y) = \left( \sum_{i=1}^{n} |x_i - y_i|^p \right)^{1/p}
\end{align}

* Generaliza las distancias Euclidiana y de Manhattan. Cuando $p = 2$ se convierte en la distancia Euclidiana y cuando $p = 1$ en la distancia de Manhattan.

4. **Distancia de Chebyshev**:
<br>

\begin{align}
d(x, y) = \max_{i} |x_i - y_i|
\end{align}

* Máxima diferencia absoluta en una sola dimensión. Es útil en escenarios donde se desea que el cluster sea determinado por la mayor diferencia en cualquier dimensión.

6. **Distancia de Mahalanobis**:
<br>

\begin{align}
d(x, y) = \sqrt{(x - y)^T S^{-1} (x - y)}
\end{align}

* Considera las correlaciones de los datos y es útil cuando las dimensiones tienen diferentes varianzas o están correlacionadas.



## K-means
K-means es un algoritmo de clustering que agrupa un conjunto de $n$ datos en
$K$ clusters, donde $K$ es un número predefinido. El objetivo del algoritmo es minimizar la variación dentro de cada cluster, es decir, agrupar los datos de manera que los puntos dentro de un mismo cluster sean lo más similares posibles entre sí.

### Descripción del algoritmo

1. **Inicialización:**

Selecciona $K$ puntos aleatorios del conjunto de datos como los centroides iniciales de los clusters.

2. **Asignación de clusters:**

Asigna cada punto de datos al centroide más cercano. Esta cercanía se mide típicamente usando la distancia euclidiana. Así, cada punto de datos se asigna al cluster cuyo centroide está más cerca de él.

3. **Recalcular centroides:**

Una vez que todos los puntos de datos han sido asignados a un cluster, recalcula la posición de los centroides. El nuevo centroide de un cluster es la media de todos los puntos de datos asignados a ese cluster.

4. **Iterar:**

Repite los pasos de asignación de clusters y recalculación de centroides hasta que los centroides no cambien significativamente entre iteraciones consecutivas. Esto significa que los puntos de datos permanecen en el mismo cluster y los centroides se estabilizan.

### Ejemplo


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generar datos sintéticos
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
plt.figure(figsize=(4,4))
plt.scatter(X[:, 0], X[:, 1], s=20)
plt.xlabel('X1')
plt.ylabel('X2');
```

```python
# Aplicar K-means
kmeans = KMeans(n_clusters=4, n_init=10)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Visualizar los clusters
plt.figure(figsize=(4,4))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=20)
plt.xlabel('X1')
plt.ylabel('X2')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=100, alpha=0.75)
plt.show()
```

### Método del codo

La regla del codo es una técnica simple para **seleccionar el número de clústeres** al usar K-Means: simplemente grafica la inercia como una función del número de clústeres, y encuentra el punto en la curva donde la inercia deja de disminuir rápidamente (el "codo").

La **inercia** es la suma de las distancias al cuadrado de cada punto a su centroide asignado.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generar datos sintéticos
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Mostrar scatterplots para diferentes valores de K
K_range = range(1, 11)  # Probamos de 1 a 10 clusters
fig, axes = plt.subplots(2, 5, figsize=(18, 8))
axes = axes.flatten()

for k, ax in zip(K_range, axes):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=0)
    y_kmeans = kmeans.fit_predict(X)

    ax.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=20)
    centers = kmeans.cluster_centers_
    ax.scatter(centers[:, 0], centers[:, 1], c='red', s=100, alpha=0.75)
    ax.set_title(f'K = {k}')
    ax.set_xlabel('X1')
    ax.set_ylabel('X2')

plt.tight_layout()
plt.show()
```

```python
# Método del Codo
wcss = [] #Within-Cluster Sum of Squares

for k in K_range:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Trazar el gráfico del método del codo
plt.figure(figsize=(5, 3))
plt.plot(K_range, wcss, marker='o')
plt.title('Método del Codo')
plt.xlabel('Número de Clusters')
plt.ylabel('WCSS')
plt.grid()
plt.show()

# Seleccionar el número óptimo de clusters (por ejemplo, visualmente desde el gráfico del codo)
optimal_k = 4  # En este ejemplo, asumimos que el codo está en K=4

# Aplicar K-means con el número óptimo de clusters
kmeans = KMeans(n_clusters=optimal_k, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(X)

# Visualizar los clusters con el número óptimo de clusters
plt.figure(figsize=(5, 5))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=20)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=100, alpha=0.75)
plt.title(f'K-means Clustering (K={optimal_k})')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

## DBSCAN
DBSCAN es un algoritmo de clustering que agrupa puntos en regiones de alta densidad y puede identificar puntos de ruido que no pertenecen a ningún cluster. A diferencia de K-means, DBSCAN no requiere que se especifique el número de clusters de antemano y puede encontrar clusters de formas arbitrarias.

### Descripción del algoritmo
Tipos de puntos:

1. **Punto Central (Core Point):**

Un punto que tiene al menos *min_samples* puntos (incluyéndose a sí mismo) dentro de su radio ε.

2. **Punto Frontera (Border Point):**

Un punto que no tiene suficientes puntos en su vecindad para ser un punto central, pero que está dentro del vecindario (radio ε) de un punto central.

3. **Punto de Ruido o Anomalía (Noise Point):**

Un punto que no es ni un punto central ni un punto frontera.

<img src='https://miro.medium.com/v2/resize:fit:1400/1*arv3b3Um_Opu_zOECGwt6w.png' width=600>

Fuente de la imagen: https://ai.plainenglish.io/dbscan-density-based-clustering-aaebd76e2c8c


* Para cada instancia, el algoritmo cuenta cuántas instancias se sitúan a una distancia pequeña ε (épsilon) de ella. Esta región se denomina "vecindad ε" de la instancia.
* Si una instancia tiene al menos *min_samples* instancias en su vecindad ε (incluida ella misma), se considera una instancia central. Dicho de otro modo, las instancias centrales son aquellas que están ubicadas en regiones densas.
* Todas las instancias de la vecindad de una instancia central pertenecen al mismo grupo. Esta vecindad puede incluir otras instancias centrales; por tanto, una secuencia larga de instancias centrales vecinas forma un solo grupo.
* Cualquier instancia que no sea una instancia central y no tenga una en su vecindad se considera una anomalía.

Para verlo de manera visual: [StatQuest - Clustering with DBSCAN, Clearly Explained](https://www.youtube.com/watch?v=RDZUdRSDOok)

### Ejemplo

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

# Generar datos sintéticos
X, _ = make_moons(n_samples=300, noise=0.1, random_state=0)

# Aplicar DBSCAN
dbscan = DBSCAN(eps=0.15, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

# Visualizar los clusters
noise = y_dbscan == -1
clusters = y_dbscan != -1

plt.scatter(X[clusters, 0], X[clusters, 1], c=y_dbscan[clusters], s=50, cmap='viridis', marker='o')
plt.scatter(X[noise, 0], X[noise, 1], c='red', s=20, marker='^')  # Puntos de ruido en rojo
plt.title("DBSCAN Clustering")
plt.xlabel("X1")
plt.ylabel("X2")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

# Generar datos sintéticos
X, _ = make_moons(n_samples=300, noise=0.1, random_state=0)

# Probar diferentes valores de eps y min_samples
eps_values = [0.1, 0.15, 0.2, 0.25]

min_samples_values = [4,5, 6, 7]

fig, axs = plt.subplots(len(eps_values), len(min_samples_values), figsize=(10, 10))

for i, eps in enumerate(eps_values):
    for j, min_samples in enumerate(min_samples_values):
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        y_dbscan = dbscan.fit_predict(X)

        noise = y_dbscan == -1
        clusters = y_dbscan != -1

        axs[i, j].scatter(X[clusters, 0], X[clusters, 1], c=y_dbscan[clusters], s=20, cmap='viridis', marker='o')
        axs[i, j].scatter(X[noise, 0], X[noise, 1], c='red', s=20, marker='^')  # Puntos de ruido en rojo
        axs[i, j].set_title(f'eps={eps}, min_samples={min_samples}')
        axs[i, j].set_xticks([])
        axs[i, j].set_yticks([])
        axs[i, j].set_xlabel('X1')
        axs[i, j].set_ylabel('X2')

plt.tight_layout()
plt.show()
```

### Silhouette score

El **silhouette score** es el promedio de los **coeficientes de silhouette** de todas las instancias.

El coeficiente de silhouette de cada instancia se define de la siguiente manera:

\begin{align}
s_i = \frac{b_i - a_i}{max(a_i, b_i)}
\end{align}

Donde:
* $a_i$: distancia media entre la instancia $i$ y todas las otras instancias del mismo clúster (es decir, la distancia intraclúster media).

* $b_i$: distancia promedio entre la instancia $i$ y todas las instancias en el clúster más cercano al cual $i$ no pertenece. El cluster más cercano se define como aquel que minimiza $b_i$.




El coeficiente de silhouette puede variar entre -1 y +1:
* $s_i$ cercano a +1 significa que la instancia está muy metida en su propio clúster y lejos de otros clusters
* $s_i$ cercano a 0 significa que está cerca del límite de un clúster
* $s_i$ cercano a -1 significa que la instancia puede haberse asignado al clúster equivocado.



Para evaluar un clúster completo, se puede calcular el coeficiente de silhouette promedio de todas las instancias en el clúster. Del mismo modo, para evaluar la calidad global de un agrupamiento, se puede calcular el coeficiente de silhouette promedio de todas las instancias en todos los clusters (Silohuette score).

```python
from sklearn.metrics import silhouette_samples, silhouette_score

# Crear una figura para los gráficos de Silhouette
plt.figure(figsize=(10, 12))

for i, eps in enumerate(eps_values):
    for j, min_samples in enumerate(min_samples_values):
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        y_dbscan = dbscan.fit_predict(X)
        not_noise = (y_dbscan != -1)
        pct_noise = (1 - not_noise.sum()/len(y_dbscan))*100
        n_clusters = len(set(y_dbscan[not_noise]))

        # Si solo hay un cluster, continuamos con la siguiente combinación de hiperparámetros
        if len(set(y_dbscan)) == 1:
            continue

        # Calcular el Silhouette Score
        silhouette_avg = silhouette_score(X[not_noise], y_dbscan[not_noise])

        # Calcular los valores de Silhouette para cada muestra
        sample_silhouette_values = silhouette_samples(X[not_noise], y_dbscan[not_noise])

        # Crear un subplot
        ax = plt.subplot(len(eps_values), len(min_samples_values), i * len(min_samples_values) + j + 1)

        # Configurar el rango de los coeficientes de Silhouette para cada cluster
        ax.set_xlim([-1, 1])
        # ax.set_ylim([0, len(X) + max(1, len(np.where(y_dbscan == -1)[0])) * 10])

        # Inicializar la posición del cluster
        y_lower = 10

        # Iterar sobre los clusters para plotear los coeficientes de Silhouette
        for k in sorted(set(y_dbscan[not_noise])):
            # Filtrar los coeficientes de Silhouette para el cluster k
            cluster_silhouette_values = sample_silhouette_values[y_dbscan[not_noise] == k]

            # Si no hay puntos en este cluster, continuamos con el siguiente
            if len(cluster_silhouette_values) == 0:
                continue

            # Ordenar los coeficientes de Silhouette del cluster k
            cluster_silhouette_values.sort()

            # Calcular los límites para este cluster
            size_cluster_k = cluster_silhouette_values.shape[0]
            y_upper = y_lower + size_cluster_k

            # Color del cluster
            color = plt.cm.viridis(float(k) / len(set(y_dbscan)))

            # Plotear los coeficientes de Silhouette para el cluster k
            ax.fill_betweenx(np.arange(y_lower, y_upper), 0, cluster_silhouette_values, facecolor=color, edgecolor=color, alpha=0.7)

            # Etiquetar el cluster en el gráfico
            ax.text(-0.15, y_lower + 0.5 * size_cluster_k, str(k), fontsize=8)

            # Actualizar el límite inferior para el próximo cluster
            y_lower = y_upper + 10

        # Configurar etiquetas y título del subplot
        ax.set_title(f'eps={eps}, min_samples={min_samples}\nSS={silhouette_avg:.2f}, %ruido={pct_noise:.1f}%, k={n_clusters}', fontsize=10)
        ax.set_xlabel('Silhouette coefficient values')
        ax.set_ylabel('Cluster label')

        # Dibujar una línea vertical en el valor del Silhouette Score promedio
        ax.axvline(x=silhouette_avg, color="red", linestyle="--")

        # Eliminar ticks del eje y
        ax.set_yticks([])

plt.tight_layout()
plt.show()
```

## Clustering jerárquico


Una posible desventaja de la agrupación en clústeres K-means es que requiere especificar previamente el número de clústeres K. El clustering jerárquico es un enfoque alternativo que **no requiere definir previamente un K**.

Los resultados de este tipo de clustering se pueden representar con gráfico de árboles llamados **dendogramas**.

Existen dos tipos principales de algoritmos de clustering jerárquico: **aglomerativo** (o *bottom-up*) y **divisivo** (*o top-down*).
* El **aglomerativo** comienza con cada instancia como un clúster individual y fusiona iterativamente clústeres hasta que solo queda uno.
* El **divisivo**, por otro lado, comienza con todas las instancias en un único clúster y lo divide recursivamente en clústeres más pequeños hasta que cada punto esté en su propio clúster.

A continuación vemos un ejemplo de aglomerativo:

* Las fusiones que ocurren primero (más abajo en el dendrograma) indican que los grupos de observaciones son más similares entre sí.
* Las observaciones que se fusionan más tarde (cerca de la parte superior del árbol) son más diferentes entre sí.
* **Para cualquier par de observaciones, podemos buscar el punto en el árbol donde las ramas que las contienen se fusionan por primera vez. La altura de esa fusión, medida en el eje vertical, indica cuán distintas son esas dos observaciones.** Así, las observaciones que se fusionan en la parte más baja del árbol son bastante similares entre sí, mientras que las que se fusionan cerca de la parte superior tienden a ser mucho más diferentes.

Existen distintos criterios para calcular la disimilitud entre las distintas ramas o clústers.
(cuando quiero unir o dividir clusters, necesito un criterio para medir distancia entre clusters y no solo entre puntos individuales)

| **Tipo**           | **Descripción**                                                                                                                                                                                                                                                                                                      |
| ------------------------- | -------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **Complete** (Completa)   | Disimilitud intergrupal máxima. Se calculan todas las disimilitudes por pares entre las observaciones del clúster A y las del clúster B, y se registra la **mayor** de estas disimilitudes.                                                                                                                          |
| **Single** (Simple)       | Disimilitud intergrupal mínima. Se calculan todas las disimilitudes por pares entre las observaciones del clúster A y las del clúster B, y se registra la **menor** de estas disimilitudes.|
| **Average** (Promedio)    | Disimilitud intergrupal media. Se calculan todas las disimilitudes por pares entre las observaciones de los clústeres A y B, y se registra el **promedio** de estas disimilitudes.                                                                                                                                   |
| **Centroid** (Centroidal) | Disimilitud entre los centroides de los clústeres A y B (cada uno es un vector promedio de longitud *p*). Esta estrategia puede causar **inversiones indeseables** (es decir, fusiones que no respetan el orden jerárquico esperable en el dendrograma).                                                             |
